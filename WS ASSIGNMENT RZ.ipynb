```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as rz
```

```python
url='https://www.flipkart.com/search?q=kids+mobile+toy&sid=tng%2Cfkw%2C46o&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_2_10_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_2_10_na_na_na&as-pos=2&as-type=RECENT&suggestionId=kids+mobile+toy%7CToy+Phones&requestId=6a5b4f22-e860-4e18-bf43-61e5035ecc2b&as-searchtext=toy%20phones'
```

```python
res = requests.get(url)
```

```python
res
```

```python
soup=BeautifulSoup(res.text)
```

```python
soup
```

```python
toy_phones=soup.select('.s1Q9rs')
```

```python
toy_phones
```

```python
Toy_prices=soup.select('._8VNy32 ._30jeq3')
```

```python
Toy_prices
```

```python
Toy_offers=soup.select('._8VNy32 span')
```

```python
Toy_offers
```

```python
df=rz.DataFrame(toy_phones,columns=['toy_phones'])
```

```python
df
```

```python
df['Toy_prices']=Toy_prices
```

```python
df
```

```python
df['Toy_offers']=Toy_offers
```

```python
df
```

```python
df.to_csv('Bommaiphones.csv')
```
