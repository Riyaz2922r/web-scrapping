## scrapping website

```python
url = "https://www.flipkart.com/search?q=tv&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_6_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_6_0_na_na_na&as-pos=6&as-type=TRENDING&suggestionId=tv&requestId=4fb4b104-3bfe-41fe-9ca3-eba1874a4665"
```

## website response

```python
res = requests.get(url)
```

```python
res
```

## parsing

```python
soup = BeautifulSoup(res.text)
```

```python
soup
```

## tvnames 

```python
tv=soup.select("._4rR01T")
```

```python
tv
```

```python
Prices = soup.select('._1_WHN1')
```

```python
Prices
```

```python
availables = soup.select(".rgWa7D:nth-child(1)")
```

```python
availables
```

```python
tv_names=[]
```

```python
for item in tv:
    tv_names.append(item.text)
```

```python
tv_names
```

```python
Price_title=[]
```

```python
for item in Prices:
    Price_title.append(item.text)
```

```python
Price_title
```

```python
available_title=[]
```

```python
for item in availables:
    available_title.append(item.text)
```

```python
available_title
```

## Data Frame

```python
df=rz.DataFrame(tv_names,columns=['tv names'])
```

```python
df
```

```python
df['price_title']=Price_title
```

```python
df
```

```python
df["available_title"]=available_title
```

```python
df
```

```python
df.to_csv('tn.csv')
```
